# Evaluation on test data

### Regression Evaluation Metrics

**Mean Absolute Error** (MAE) is the mean of the absolute value of the errors:

$$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$

**Mean Squared Error** (MSE) is the mean of the squared errors:

$$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$

**Root Mean Squared Error** (RMSE) is the square root of the mean of the squared errors:

$$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

**Q:What difference will it make if I use ravel() function instead of flatten()?**
**ANSWER**
Both flatten() and ravel() are used to convert multi-dimensional arrays into one-dimensional arrays, but they have a subtle difference. flatten() returns a new array with a copy of the data, while ravel() returns a flattened view of the original array whenever possible, which means it might share data with the original array without making a copy.

In the context of your code, if you use ravel() instead of flatten(), it may still work because the flattened view should be compatible with the fully connected layers. However, you need to be cautious when modifying the original data after using ravel() because changes might affect both the original and the flattened view.


# **Question:** What is batch normalization and what ouput will be generated from the above line?
**ANSWER**
Batch normalization is a technique used in neural networks to improve the training speed and stability of a model. It works by normalizing the input of a layer before applying the activation function. The primary purpose of batch normalization is to address the internal covariate shift problem, which refers to the change in the distribution of activations in hidden layers during training.

The output will be that the code will dd batch normalization layers after two convolutional layers. After the line x = BatchNormalization()(x), the variable x will contain the normalized activations from the previous convolutional layer.

# **Question:** What type of pooling is applied?

MaxPooling2D is applied. Max pooling is a pooling operation that selects the maximum value from a group of values within a defined window (in this case, a 2x2 window) and reduces the spatial dimensions of the input while retaining the most important information.
x = MaxPooling2D((2, 2))(x) applies max pooling to the previous layer x.

#**Question:** Why is the data flattened?

**ANSWER:**
The data is flattened to transform the 2D feature maps obtained from the convolutional layers into a 1D vector. This is typically done before passing the data to fully connected (dense) layers in the neural network. Flattening is necessary because fully connected layers expect a 1D input, and it helps in connecting all the features to the neurons in the next layers.
x = Flatten()(x) flattens the 2D feature maps into a 1D vector.

#**Question:** What if we add dropout before adding the hidden layer?

**ANSWER:**
Adding dropout before a hidden layer means applying dropout to the flattened output before it's passed to the dense (hidden) layer. Dropout randomly sets a fraction of the input units to zero during each training batch, which helps prevent overfitting by reducing co-dependency between neurons. It introduces some level of regularization to the network. It's a common practice to apply dropout before fully connected layers (hidden layers) to prevent overfitting.


#**Question:** What does dropout do in a hidden layer?
**ANSWER:**
Dropout is a regularization technique used during training in neural networks. It works by randomly deactivating (setting to zero) a certain percentage of neurons (or units) in a layer during each training batch. This prevents neurons from relying too heavily on any specific neurons in the previous layer, effectively reducing co-dependency. Dropout helps in preventing overfitting by making the network more robust and less prone to memorizing the training data. During inference (prediction), dropout is typically turned off.

#**Question:** What is data augmentation?
**ANSWER:**
Data augmentation is a technique used in machine learning and deep learning to artificially increase the size of a training dataset by applying various transformations to the existing data. These transformations can include operations like rotation, translation, scaling, flipping, cropping, and more. Data augmentation is primarily used in computer vision tasks, such as image classification, to enhance the model's ability to generalize to different variations of the same data. By exposing the model to augmented data during training, it becomes more robust and less prone to overfitting.

#**Question:** What happens if we run this AFTER calling the previous model.fit()?

**ANSWER:**
If we run data augmentation after calling model.fit(), the data augmentation will have no effect on the training process for the current training session. The model.fit() function processes the training data as it is at the time of the call and trains the model based on that data. Any subsequent changes to the data, such as data augmentation, will not impact the current training session.

To apply data augmentation, it should be incorporated before training the model, typically by using data generators like ImageDataGenerator during model training setup. This way, the model is trained on augmented data from the beginning of the training process.
